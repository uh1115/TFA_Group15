```python
import pandas as pd
```

```python
df = pd.read_csv('311_Service_Requests_2020.csv')
df
```

```python
df_zip = df[df['Incident Zip'] == 10025.0]
df_zip
```

```python
df_zip.info()
```

```python
df_zip['Complaint Type'].unique()
```

```python
df_illegal_zipcode = df_zip.groupby('Complaint Type').size().loc['Illegal Parking']
df_illegal_zipcode
```

```python
df_illegal_zipcode_1 = (df_zip.groupby('Complaint Type').size().loc['Illegal Parking'] / len(df_zip.index)) * 100
df_illegal_zipcode_1
```

```python
df_illegal_total = df.groupby('Complaint Type').size().loc['Illegal Parking']
df_illegal_total
```

```python
df_illegal_total_1 = (df.groupby('Complaint Type').size().loc['Illegal Parking'] / len(df.index)) * 100
df_illegal_total_1
```

```python
higher_parking_proportion = df_illegal_zipcode_1 > df_illegal_total_1
higher_parking_proportion
```
