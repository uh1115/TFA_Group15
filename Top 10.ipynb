```python
# importing data from csv file and importing required Python libries

import pandas as pd
import numpy as np
df = pd.read_csv("311_Service_Requests_2020.csv")
```

```python
# viewing the dataframe from the csv file
df
```

```python
# checking the basic info of datatype for each column in the dataframe
df.info()
```

```python
# choose zip code as "10025"
df = df[df['Incident Zip'] == 10025.0]
# converting the zip code from float into an integer
df['Incident Zip'] = df['Incident Zip'].astype(int)

# choose top 10 incidents of calls in 10025 by grouping the "Complaint Type" in descending order
top10 = df.groupby('Complaint Type').size().sort_values(ascending = False)
top10.head(10)
```
